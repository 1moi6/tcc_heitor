### Documentação das Decisões de Design (Semana 2)

Seguindo as orientações do roteiro de trabalho, foram estabelecidas as seguintes premissas de design para a implementação da Cifra de Hill:

**1. Tratamento de Caracteres Especiais e Pontuação** 
* **Decisão:** O sistema adotará a **normalização para o alfabeto A-Z (26 letras)**.
* **Justificação:** Para manter a fidelidade ao modelo matemático sobre $\mathbb{Z}_{26}$, todos os caracteres que não pertençam ao alfabeto (espaços, pontuação, números) serão removidos.
* **Acentos:** Caracteres acentuados serão convertidos para a sua forma base (ex: 'Ã' torna-se 'A') antes do processamento. Esta abordagem simplifica a aritmética modular sem comprometer a demonstração da cifra.

**2. Definição do Tamanho de Bloco ($n$)** 
* **Decisão:** Foi selecionado um **tamanho de bloco $n=3$**.
* **Justificação:** O valor $n=3$ oferece um equilíbrio ideal para o projeto:
    * É suficientemente pequeno para permitir cálculos manuais de verificação durante o desenvolvimento.
    * É suficientemente grande para demonstrar a **difusão** (uma letra do texto claro afeta três no cifrado) e para tornar o ataque por texto claro conhecido mais interessante do ponto de vista algébrico.
    * Em caso de mensagens cujo comprimento não seja múltiplo de 3, será aplicado um *padding* com o caractere 'X'.

**3. Estratégia para Geração de Chaves Invertíveis** 
* **Decisão:** Utilização de **sorteio aleatório com critério de rejeição**.
* **Justificação:** A chave é uma matriz $A \in M_{3}(\mathbb{Z}_{26})$. Para garantir a decifragem, a matriz deve ser obrigatoriamente invertível. O algoritmo implementado gera valores aleatórios entre 0 e 25 para cada posição da matriz e valida se $\gcd(\det A, 26) = 1$. Se a condição não for satisfeita, a matriz é descartada e uma nova é gerada até que se obtenha uma chave válida.
