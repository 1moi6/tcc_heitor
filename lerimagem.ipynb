```python
import numpy as np
import PIL
from PIL import Image
from numpy import asarray
import matplotlib.pyplot as plt
```

```python
im = Image.open('cofe.png').convert('L')
data = asarray(im)
# id = 1
# plt.plot(data[300,:]>250)
# plt.plot(data[:,id,1])
# plt.plot(data[:,id,2])
# gr_im= Image.fromarray(data).save('gr_kolala.png')
# im = plt.imread("cofe.png",)
data = asarray(im)
# plt.imshow(data<28)

# x = np.linspace(0,1,data.shape[0])
# y = np.linspace(0,1,data.shape[1])
vx,vy=[],[]
for z in range(data.shape[1]):
    xx = list(data[:,z]<28)
    xx = [int(v) for v in xx]
    xx =[ i for i,e in enumerate(xx) if e==1]
    vy.append(1-np.mean(xx)/data.shape[0])
    vx.append(z/data.shape[1])
```

```python
ax = plt.plot(vx,vy)
plt.xlim(0,1)
plt.ylim(0,1)
```
