```python
import numpy as np
import PIL
from PIL import Image
from numpy import asarray
import matplotlib.pyplot as plt
import os
files = os.listdir(os.getcwd())
files = sorted([file for file in files if file.endswith(".png")])
print(files)

```

```python
import json
def classificacao(parametro,leitura):
    with open('classificacao_iqa.json', 'r') as f:
        dados_json = f.read()
    classi = json.loads(dados_json)

    vleitura = np.array(classi[parametro]["leitura"])
    idx = np.argmin((vleitura-leitura)**2)
    cl = classi[parametro]["classificacao"][idx]

    return cl

```

```python
classificacao("ecoli",5000)
```
