```python
import numpy as np
import PIL
from PIL import Image
from numpy import asarray
import matplotlib.pyplot as plt
import os
files = os.listdir(os.getcwd())
files = sorted([file for file in files if file.endswith(".png")])
```

```python
im = Image.open(files[0]).convert('L')
data = asarray(im)

data = asarray(im)
data2 = data[2:-2,2:-2]
vx,vy=[],[]
for z in range(data2.shape[1]):
    xx = list(data2[2:-2,z]<28)
    xx = [int(v) for v in xx]
    xx =[ i for i,e in enumerate(xx) if e==1]
    vy.append(1-np.mean(xx)/data2.shape[0])
    vx.append(z/data2.shape[1])
```

```python
ax = plt.plot(vx,vy)
plt.xlim(0,1)
plt.ylim(0,1)
```

```python
vxm = [10**(v*10/2) for v in vx] 
fig, ax = plt.subplots()
ax.semilogx(vxm,[100*v for v in vy])
ax.set_ylim([0,100])
ax.set_xlim([1,10**5])
```
